```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({
    "size":[
        's','m','l','xl','xxl',
        'xxl','l','l','m','xxl'
    ]
})
df.head(6)
```

```python
ord_data=[['s','m','l','xl','xxl']]
```

```python
oe=OrdinalEncoder(categories=ord_data)
oe.fit(df[['size']])
```

```python
df['en_size']=oe.transform(df[['size']])
df.head(6)
```

```python
# Using map function
ord_data_map={'s':0,
             'm':1,
             'l':2,
             'xl':3,
             'xxl':4}
```

```python
df['en_map_size']=df['size'].map(ord_data_map)
df.head(6)
```
