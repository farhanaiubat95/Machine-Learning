```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset=pd.read_csv("diabetes.csv")
dataset.head(4)
```

```python
# x & y axis

x=dataset.iloc[:,:-1]
y=dataset["Outcome"]
```

```python
x.shape
```

```python
# depend on this dataset - need a model

from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=10000)
```

```python
fs=SequentialFeatureSelector(lr,k_features=5,forward=False,scoring='accuracy')
fs.fit(x,y)
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
