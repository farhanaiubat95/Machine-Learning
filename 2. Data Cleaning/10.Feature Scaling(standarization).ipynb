```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# By using real data - sklearn
dataset=pd.read_csv("student_habits_performance.csv")
dataset.head(6)
```

```python
dataset.isnull().sum()
```

```python
# dataset["netflix_hours"].fillna(dataset["netflix_hours"].mean())
```

```python
sns.distplot(dataset["netflix_hours"])
plt.show()
```

```python
dataset.describe()
```

```python
# using scikit learn

ss=StandardScaler()
ss.fit(dataset[["netflix_hours"]])
```

```python
ss_transData=ss.transform(dataset[["netflix_hours"]])
ss_transData
```

```python
dataset["netflix_hours_ss"]=pd.DataFrame(ss_transData,columns=["x"])
dataset.head(6)
```

```python
# nature check -> same or different
plt.figure(figsize=(12,5))
plt.subplot(1,2,1) # (row, col, position)
plt.title("Before")
sns.distplot(dataset["netflix_hours"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["netflix_hours_ss"])
plt.show()
```
