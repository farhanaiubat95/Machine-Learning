```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# By using real data - sklearn
dataset=pd.read_csv("student_habits_performance.csv")
dataset.head(6)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
sns.boxplot(x="netflix_hours",data=dataset)
plt.show()
```

```python
dataset.shape
```

```python
q1=dataset["netflix_hours"].quantile(0.25)
q3=dataset["netflix_hours"].quantile(0.75)
print(q1, q3)
```

```python
IQR= q3-q1
print(IQR)
```

```python
min_range=q1-(1.5*IQR)
max_range=q3+(1.5*IQR)
print(min_range, max_range)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x="netflix_hours",data=dataset)
plt.show()
```

```python
dataset
```

```python
new_dataset=dataset[dataset["netflix_hours"]<=max_range]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="netflix_hours",data=new_dataset)
plt.show()
```
