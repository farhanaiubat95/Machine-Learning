```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# By using real data - sklearn
dataset=pd.read_csv("student_habits_performance.csv")
dataset.head(6)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["netflix_hours"])
plt.show()
```

```python
# using scikit learn

ms=MinMaxScaler()
ms.fit(dataset[["netflix_hours"]])
```

```python
ms_transData=ms.transform(dataset[["netflix_hours"]])
```

```python
dataset["netflix_hours_ms"]=ms_transData
dataset.head(6)
```

```python
# nature check -> same or different
plt.figure(figsize=(12,5))
plt.subplot(1,2,1) # (row, col, position)
plt.title("Before")
sns.distplot(dataset["netflix_hours"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["netflix_hours_ms"])
plt.show()
```
