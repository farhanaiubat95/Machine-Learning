```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.DataFrame({
    "name":[
        "farhana",
        "pen",
        "paper",
        "food",
        "bag"
    ]
})
df
```

```python
# Using Scikit-Learn
le=LabelEncoder()
df["en_name"]=le.fit_transform(df["name"])
df
```
