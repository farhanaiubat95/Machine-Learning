```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset=pd.read_csv("placement.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=dataset)

# Regression line (straight line fit)
sns.regplot(x="cgpa", y="package", data=dataset, scatter=False, color='red')
plt.title("CGPA vs Package")
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.tight_layout() # adjusts layout to avoid overlap.
plt.show()
```

```python
# input & output separate

x=dataset[["cgpa"]] # -> 2D
y=dataset["package"] # -> 1D
x.ndim
```

```python
# Train & Test

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42) # random_state=42 -> ideal number , test_size=0.2 -> 20%
```

```python
# Train Model
# Linear equation : y = m*x+c
# Unknown : m & c ( find out m & c for calculation)

lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
# let's check our prediction value manually

m = lr.coef_
c = lr.intercept_
m , c
```

```python
y = m * 6.89 + c
y
```

```python
# predict & test

predict_value = lr.predict([[6.89]])
predict_value
```

```python
# manual value & train model value is same or diff

y==predict_value
```

```python
y_prd=lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="package",data=dataset)

# Regression line (straight line fit)
plt.plot(dataset["cgpa"], y_prd, color='red')

# Title
plt.title("Simple Linear Regression")

# Line & dot
plt.legend(["Original Data","Predict Line"])

# Save image
plt.savefig("1. Predict-Simple Linear Regression .jpg")

plt.show()
```
